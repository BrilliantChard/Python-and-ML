# ***1. Line Plot***

```python
plt.figure(figsize=(14, 6))

sns.lineplot(data=insurance_data['age'])
sns.lineplot(data=insurance_data['bmi'])

plt.xlabel(' ')
plt.title('Line Plot showing the relationship between Age and BMI')
```

# ***Scatter Plot***

```python
insurance_data.columns
```

```python
plt.figure(figsize=(14, 6))

sns.scatterplot(x='age', y='bmi', hue='smoker', data=insurance_data)

plt.title('Scatter plot showing Relationship between Age and BMI')
```

```python
insurance_data.columns
```

```python
plt.figure(figsize=(14, 6))

sns.scatterplot(x='bmi', y='charges', hue='smoker', data=insurance_data)

plt.title('Scatter plot showing Relationship between BMI and Charges')
```

```python
plt.figure(figsize=(14, 6))

sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)

plt.title('Scatter plot showing Relationship between BMI and Charges')
```

```python
plt.figure(figsize=(14, 6))

sns.swarmplot(x='bmi', y='charges', hue='smoker', data=insurance_data, alpha=0.7)

plt.title('Scatter plot showing Relationship between BMI and Charges')
plt.show()
```
